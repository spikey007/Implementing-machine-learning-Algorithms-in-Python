```python
def chief_string(title):
    if  title.strip() == 'United-States':
        return(title)
    else:
        return("others")
```

# combining some of the levels in Working sector column

```python
income["working_sector"].value_counts()
```

```python
def working(rcb):
    if rcb.strip() in ("state","national","local_body"):
        return('government')
    elif rcb.strip() in ("without_pay","not_worked"):
        return('not working')
    elif rcb.strip() in ("private","self_employed"): 
        return('private')
 
        
```

```python
income["working_sector"]=(income["working_sector"].apply(lambda v:working(v)))
```

```python
income["working_sector"].value_counts()
```

# combining some of the levels in qualification column

```python
income["qualification"].value_counts()
```

```python
def qual(bulls):
    if bulls.strip() in ("10th","1st-4th","5th-6th","9th","7th-8th"):
        return("school")
    if bulls.strip() in ("11th","12th"):
        return("collage")
    else:
        return(bulls)
```

```python
income["qualification"]=(income["qualification"].apply(lambda r:qual(r)))
```

```python
income["qualification"].value_counts()
```

```python
def mart(fc):
    if fc in ("Divorced","Separated","Widowed","Never-married"):
        return("not married")
    else:
        return("married")
    
```

# Use One-Hot encoder to encode categorical variables

```python
from sklearn import preprocessing
```

```python
encoder = preprocessing.LabelEncoder()
```

```python
income_copy=income
```

```python
#fit and transform all known categorical variables 
working_sector_cat = encoder.fit_transform(income_copy.working_sector)
occupation_cat = encoder.fit_transform(income_copy.occupation)
qualification_cat = encoder.fit_transform(income_copy.qualification)
country_cat = encoder.fit_transform(income_copy.country)
maritial_cat = encoder.fit_transform(income_copy.marital_status)
ethnicity_cat = encoder.fit_transform(income_copy.ethnicity)
rel_cat = encoder.fit_transform(income_copy.relationship)
gender_cat = encoder.fit_transform(income_copy.gender)
```

```python
#Now initialze the cols with encoded values
income_copy['working_sector_cat']=working_sector_cat
income_copy['occupation_cat']=occupation_cat
income_copy['qualification_cat']=qualification_cat
income_copy['country_cat']=country_cat
income_copy['maritial_cat']=maritial_cat
income_copy['ethnicity_cat']=ethnicity_cat
income_copy['rel_cat']=rel_cat
income_copy['gender_cat']=gender_cat
```

```python
#Now drop all old categorical variables
droplist = ['working_sector','qualification','marital_status','occupation','relationship','ethnicity','gender','country']
income_copy = income_copy.drop(droplist, axis=1)
```

```python
 from sklearn.model_selection import train_test_split
```

```python
from sklearn.model_selection import train_test_split

Y = income_copy ['target']
X = income_copy.drop('target', axis=1)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=0.33, random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression

model_logistic = LogisticRegression()

model_logistic.fit(X_train, Y_train)
Y_pred_logistic = model_logistic.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, recall_score, classification_report,confusion_matrix

```

```python
#check performance
print ("Classification Report:")
print(classification_report(Y_test, Y_pred_logistic))
print ("Confusion Matrix:")
print(confusion_matrix(Y_test, Y_pred_logistic))
print ("Recall Score:")
print(recall_score(Y_test, Y_pred_logistic))
print ("Accuracy Score:")
print(accuracy_score(Y_test, Y_pred_logistic))
```

# Naive Bayes algorithm
Naive Bayes uses the Bayes theorem to calculate the probability of a given feature set falls into a certain class or not. We will use the Gaussian NB method

# SVM
Use the SVM classifier for the same classification problem. We can try different C values and kernels to get a better result

```python
from sklearn import svm
from sklearn.model_selection import GridSearchCV
```

```python
model_svm = svm.SVC(C=10, kernel='sigmoid')
model_svm.fit(X_train, Y_train)

#predict on Test
Y_pred_svm = model_svm.predict(X_test)
```

```python
#check performance
print ("Classification Report:")
print(classification_report(Y_test, Y_pred_svm))
print ("Confusion Matrix:")
print(confusion_matrix(Y_test, Y_pred_svm))
print ("Recall Score:")
print(recall_score(Y_test, Y_pred_svm))
print ("Accuracy Score:")
print(accuracy_score(Y_test, Y_pred_svm))
```

```python
#Create a dictionary of possible parameters
params_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100],
          'gamma': [0.0001, 0.001, 0.01, 0.1],
          'kernel':['linear','rbf'] }

#Create the GridSearchCV object
grid_clf = GridSearchCV(svm.SVC(class_weight='balanced'), params_grid)

#Fit the data with the best possible parameters
grid_clf.fit(X_train, Y_train)

#Print the best estimator with it's parameters
print (grid_clf.best_params_)
```

# SVM - results of hyperparameter optimization
The results of GridSearch in SVM is:

{'C': 0.1, 'gamma': 0.0001, 'kernel': 'rbf'}

Let's try the model based on this

```python
model_svm_optimal = svm.SVC(C=0.1, kernel='sigmoid',gamma=0.0001)
model_svm_optimal.fit(X_train, Y_train)

#predict on Test
Y_pred_svm_opt = model_svm_optimal.predict(X_test)
```

```python
#check performance
print ("Classification Report:")
print(classification_report(Y_test, Y_pred_svm_opt))
print ("Confusion Matrix:")
print(confusion_matrix(Y_test, Y_pred_svm_opt))
print ("Recall Score:")
print(recall_score(Y_test, Y_pred_svm_opt))
print ("Accuracy Score:")
print(accuracy_score(Y_test, Y_pred_svm_opt))
```

# Tree value of 120 is giving a good accuracy of 85%
Use k-fold cross validation to improve accuracy
Use k-fold cross validation to see if it can improve the accuracy for all algorithms

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import GridSearchCV
```

```python
rf_class_kfold = RandomForestClassifier(n_estimators=120)
log_class_kfold = LogisticRegression()
svm_class_kfold = svm.SVC()

number_of_folds=10

print("Random Forests: ")
print(cross_val_score(rf_class_kfold, X_train, Y_train, scoring='accuracy', cv = number_of_folds))
accuracy = cross_val_score(rf_class_kfold, X_train, Y_train, scoring='accuracy', cv = number_of_folds).mean() * 100
print("Accuracy of Random Forests using K-fold cross validation is: " , accuracy)
 
print("\n\nSVM:")
print(cross_val_score(svm_class_kfold, X_train, Y_train, scoring='accuracy', cv = number_of_folds))
accuracy = cross_val_score(svm_class_kfold, X_train, Y_train, scoring='accuracy', cv = number_of_folds).mean() * 100
print("Accuracy of SVM using K-fold cross validation is: " , accuracy)
 
print("\n\nLog:")
print(cross_val_score(log_class_kfold, X_train, Y_train, scoring='accuracy', cv = number_of_folds))
accuracy = cross_val_score(log_class_kfold, X_train, Y_train, scoring='accuracy', cv = number_of_folds).mean() * 100
print("Accuracy of Logistic Regression using K-fold cross validation is: " , accuracy)
```
